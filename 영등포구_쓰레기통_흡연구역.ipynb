```python
df_a = pd.read_csv('https://raw.githubusercontent.com/ecothonCBB/CigaretteButtBins/main/data/%E1%84%89%E1%85%A5%E1%84%8B%E1%85%AE%E1%86%AF%E1%84%90%E1%85%B3%E1%86%A8%E1%84%87%E1%85%A7%E1%86%AF%E1%84%89%E1%85%B5%20%E1%84%8B%E1%85%A7%E1%86%BC%E1%84%83%E1%85%B3%E1%86%BC%E1%84%91%E1%85%A9%E1%84%80%E1%85%AE_%E1%84%92%E1%85%B3%E1%86%B8%E1%84%8B%E1%85%A7%E1%86%AB%E1%84%80%E1%85%AE%E1%84%8B%E1%85%A7%E1%86%A8(%E1%84%89%E1%85%B5%E1%86%AF%E1%84%8B%E1%85%AC)%20%E1%84%92%E1%85%A7%E1%86%AB%E1%84%92%E1%85%AA%E1%86%BC_20220429.csv', encoding='CP949')
df_a.head()
```

```python
df_a[['위도','경도']].duplicated().sum()
```

```python
df_a.drop_duplicates(subset=['위도','경도'], inplace=True)
df_a[['위도','경도']].duplicated().sum()
```

```python
import folium

# 영등포구의 위도
latitude = 37.52361111
longitude = 126.8983417

yd = folium.Map(location=[latitude, longitude],
               zoom_start=13, 
               width=1200, 
               height=700
              )
```

### 🔷 꽁초 쓰레기통, 🔴 흡연구역

```python
m_map = df_b[["위도", "경도"]].dropna()
for lat, long in zip(m_map["위도"], m_map["경도"]):
    folium.Marker([lat, long]).add_to(yd)

m_map2 = df_a[["위도", "경도"]].dropna()
for lat, long in zip(m_map2["위도"], m_map2["경도"]):
    folium.CircleMarker([lat, long], radius=23, color='#cc4831', fill_color='#cc4831').add_to(yd)

yd
```
