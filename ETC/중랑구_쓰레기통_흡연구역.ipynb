```python
jf_a = pd.read_csv('https://raw.githubusercontent.com/ecothonCBB/CigaretteButtBins/main/data/%E1%84%89%E1%85%A5%E1%84%8B%E1%85%AE%E1%86%AF%E1%84%90%E1%85%B3%E1%86%A8%E1%84%87%E1%85%A7%E1%86%AF%E1%84%89%E1%85%B5_%E1%84%8C%E1%85%AE%E1%86%BC%E1%84%85%E1%85%A1%E1%86%BC%E1%84%80%E1%85%AE_%E1%84%92%E1%85%B3%E1%86%B8%E1%84%8B%E1%85%A7%E1%86%AB%E1%84%89%E1%85%B5%E1%84%89%E1%85%A5%E1%86%AF_20221011.csv')
jf_a.head()
```

```python
address= jf_a['주소']
for i in range(len(address)):
    a = address[i].split(',')[0].split()
    address[i] = " ".join(a[0:4]).split('(')[0]

print(address)
```

```python
jf_a = jf_a.drop_duplicates(subset='주소')
jf_a['주소'].duplicated().sum()
```

```python
import json
import requests

def addr_to_lat_lon(addr):
    url = 'https://dapi.kakao.com/v2/local/search/address.json?query={address}'.format(address=addr)
    headers = {"Authorization": "KakaoAK " + '###write your key###'}
    result = json.loads(str(requests.get(url, headers=headers).text))
    match_first = result['documents'][0]['address']
    return float(match_first['y']), float(match_first['x'])

address= jf_a['주소']
print(address.head())

latitude = []
longitude =[]

for i in address:
    try:
        result = addr_to_lat_lon(i)
        latitude.append(result[0])
        longitude.append(result[1])
    except Exception as e:
        print(f"주소 검색 오류: {e}")
        latitude.append(None)  # 또는 다른 기본값 설정
        longitude.append(None)

jf_a['위도'] = latitude
jf_a['경도'] = longitude
jf_a.head()
```

```python
# 네이버에 주소/시설명으로 검색 -> 없음 -> 폐업으로 판단
jf_a[jf_a['위도'].isnull()]
```

```python
jf_a = jf_a.dropna(subset=['위도'], how='any')
jf_a['위도'].isnull().sum()
```

```python
import folium

# 중랑구의 위도
latitude = 37.60380556
longitude = 127.0947778

j_seoul = folium.Map(location=[latitude, longitude],
               zoom_start=14, 
               width=1200, 
               height=700
              )
```

### 🔷 꽁초 쓰레기통, 🔴 흡연구역

```python
m_map = jf_b[["위도", "경도"]].dropna()
for lat, long in zip(m_map["위도"], m_map["경도"]):
    folium.Marker([lat, long]).add_to(j_seoul)

m_map2 = jf_a[["위도", "경도"]].dropna()
for lat, long in zip(m_map2["위도"], m_map2["경도"]):
    folium.CircleMarker([lat, long], radius=23, color='#cc4831', fill_color='#cc4831').add_to(j_seoul)

j_seoul
```
